# Point Differential Dataset

This notebook converts `Dataset/Tournament Matchups.csv` (two rows per game) into one row per game with its corresponding point differential. 

```python
import pandas as pd

matchups = pd.read_csv("Dataset/Tournament Matchups.csv")

games = []
for (year, current_round), g in matchups.groupby(["YEAR", "CURRENT ROUND"], sort=True):
    g = g.sort_values("BY YEAR NO")
    a = g.iloc[0::2].reset_index(drop=True)  # Team 1
    b = g.iloc[1::2].reset_index(drop=True)  # Team 2

    game = pd.DataFrame(
        {
            "YEAR": year,
            "CURRENT_ROUND": current_round,
            "TEAM1_ID": a["TEAM NO"],
            "TEAM1": a["TEAM"],
            "TEAM1_SCORE": a["SCORE"],
            "TEAM2_ID": b["TEAM NO"],
            "TEAM2": b["TEAM"],
            "TEAM2_SCORE": b["SCORE"],
        }
    )
    game["POINT_DIFFERENTIAL"] = game["TEAM1_SCORE"] - game["TEAM2_SCORE"]
    games.append(game)

game_point_diffs = pd.concat(games, ignore_index=True)
game_point_diffs.head()
game_point_diffs.to_csv("Dataset/Tournament Game Point Differentials.csv", index=False)
```
